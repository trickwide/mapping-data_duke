Load the CSV file and extract the data

```python
import pandas as pd

df = pd.read_csv("sample_data/wine-ratings-small.csv", index_col=0) # read the csv file and set the index column to 0
df.head()
```

```python
variety_mask = df.loc[:,'variety'] == 'Red Wine'
rating_mask = (df.loc[:,'rating'] >= 90) & (df.loc[:,'rating'] <= 92) 
region_mask = df.loc[:,'region'].str.contains('Sierra Foothills')

filtered_wines = df[variety_mask & rating_mask & region_mask]

filtered_wines
```

To JSON

```python
with open('wine_list.json', 'w') as f:
    filtered_wines.to_json(f)
```
